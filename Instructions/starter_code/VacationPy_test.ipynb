### Store Part I results into DataFrame

```python
weatherdf = pd.read_csv("../output_data/weather_info.csv")
weatherdf.head()
```

### Humidity Heatmap
* Configure gmaps.
* Use the Lat and Lng as locations and Humidity as the weight.
* Add Heatmap layer to map.

```python
# configuring for gmaps
gmaps.configure(api_key=g_key)

# Storing latitude and longitude as locations for streamline purposes
locations = weatherdf[["Lat", "Lng"]]

# configuring humidity as weight for heat map
humidity = weatherdf["Humidity"]

# setting up parameters for figure
fig = gmaps.figure(center=(0,10), zoom_level = 1.75)

# setting up heat layer and adding figure layer
humid_layer = gmaps.heatmap_layer(locations, weights=humidity, dissipating=False, max_intensity=100, point_radius=3)
fig.add_layer(humid_layer)

# displaying figure - figure only shows when code is ran
fig
```

### Create new DataFrame fitting weather criteria
* Narrow down the cities to fit weather conditions.
* Drop any rows will null values.

```python
# Setting up parameters for "ideal" weather conditions
# Max Temp between 55-85 F, Humidity less than 65% Wind Speed less than 10 mph, Cloudiness less than 60%
ideal_df = weatherdf.loc[(weatherdf["Max Temp (F)"]<=85) & (weatherdf["Max Temp (F)"]>=55) &(weatherdf["Humidity"]<65) & \
                         (weatherdf["Wind Speed"]<10) & (weatherdf["Cloudiness"]<61) & (weatherdf["Cloudiness"]>10)].dropna()
ideal_df
```

### Hotel Map
* Store into variable named `hotel_df`.
* Add a "Hotel Name" column to the DataFrame.
* Set parameters to search for hotels with 5000 meters.
* Hit the Google Places API for each city's coordinates.
* Store the first Hotel result into the DataFrame.
* Plot markers on top of the heatmap.

```python
# setting up new variable for hotel_df
hotel_df = ideal_df

# adding new column for hotel name to hotel_df
hotel_df.reset_index(drop=True, inplace=True)
hotel_df["Hotel Name"]=""
#hotel_df - this was to test that index was reset along with adding new column

# setting parameters for hotel within 5000 meters
params = {"radius": 5000, "keyword": "hotel", "key": g_key}

# setting up loop to find hotels within params
for index, row in hotel_df.iterrows():
    lat = row["Lat"]
    lng = row["Lng"]
    params["location"] = f"{lat},{lng}"
    
    base_url = "https://maps.googleapis.com/maps/api/place/nearbysearch/json"
    request = requests.get(base_url, params=params).json()
    
    # including try/except loop to account for possible errors in finding hotels
    try:
        hotel_df.loc[index, "Hotel Name"] = request["results"][0]["name"]
    except (KeyError, IndexError):
        print("Missing Data")
        pass
hotel_df
```
